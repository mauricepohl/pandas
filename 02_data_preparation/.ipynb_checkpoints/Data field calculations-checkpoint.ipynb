```python
import pandas as pd
from datetime import datetime
import math
```

```python
data = pd.read_csv("../00_data/merged_all_by_algo.csv", sep=";", decimal=",")
data['datetime'] = pd.to_datetime(data['datetime'])
data = data.set_index('datetime')
data.head(3)
```

```python
def determineSeason(inputMonth):
    if inputMonth == 12 or inputMonth <= 2: # winter
        return 0
    elif inputMonth > 2 and inputMonth <= 5: # spring
        return 1
    elif inputMonth > 5 and inputMonth <= 8: # summer
        return 2
    elif inputMonth > 8 and inputMonth <= 11: # autumn
        return 3
    return -1 # error

data["total_generation_forecast"] = data["wind_generation_forecast"] + data["solar_generation_forecast"]
data["total_generation_actual"] = data["solar_generation_actual"]+ data["wind_generation_actual"]
data["wind_forecast_error"] = data["wind_generation_actual"] - data["wind_generation_forecast"]
data["solar_forecast_error"] = data["solar_generation_actual"] - data["solar_generation_forecast"]
data["forecast_error"] = data["total_generation_actual"] - data["total_generation_forecast"]
data.loc[(data["wind_forecast_error"]==0) & (data["wind_generation_forecast"]== 0),"wind_forecast_error_percent"] = 0
data.loc[(data["solar_forecast_error"]==0) & (data["solar_generation_forecast"]== 0),"solar_forecast_error_percent"] = 0
data["pricediff_percent"] = data["price_diff"]/data["day_ahead_price"]
print(data.head(5))
# all moving averages are calculated with exponential smoothing
# moving average 30 days of forecast error
data["mov_avg_30_forecast_error"] = data.forecast_error.ewm(span=30*24,  min_periods=1).mean()
data["mov_avg_30_wind_forecast_error"] = data.wind_forecast_error.ewm(span=30*24,  min_periods=1).mean()
data["mov_avg_30_solar_forecast_error"] = data.solar_forecast_error.ewm(span=30*24,  min_periods=1).mean()
# mov avg 10 days
data["mov_avg_10_forecast_error"] = data.forecast_error.ewm(span=10*24,  min_periods=1).mean()
data["mov_avg_10_wind_forecast_error"] = data.wind_forecast_error.ewm(span=10*24,  min_periods=1).mean()
data["mov_avg_10_solar_forecast_error"] = data.solar_forecast_error.ewm(span=10*24,  min_periods=1).mean()
# forecast error avg 24 hours
data["mov_avg_1_forecast_error"] = data.forecast_error.ewm(span=24,  min_periods=1).mean()
data["mov_avg_1_wind_forecast_error"] = data.wind_forecast_error.ewm(span=24,  min_periods=1).mean()
data["mov_avg_1_solar_forecast_error"] = data.solar_forecast_error.ewm(span=24,  min_periods=1).mean()
# price avg 24 hours
data["mov_avg_1_day_ahead"] = data.day_ahead_price.ewm(span=24,  min_periods=1).mean()
data["mov_avg_1_intraday"] = data.intraday_price.ewm(span=24,  min_periods=1).mean()

# season and weekday calculation
data["season"] = data.index.month.map(lambda x: determineSeason(x))
data["month"] = data.index.month
data["weekday"] = data.index.weekday
print(data.head(35))

data.to_csv("../00_data/data_with_calc_fields.csv", sep=";", decimal=",")
print("done")
```

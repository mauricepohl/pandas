```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima_model import ARIMA
```

```python
data = pd.read_csv("../00_data/data_with_calc_fields.csv", sep=";", decimal=",")

start = 11111
stop = 11831

x = data["datetime"].iloc[start:stop]
```

```python
# wind-, solar-, and total forecast plotted vs. dateTime

yWind = data["wind_generation_forecast"].iloc[start:stop]
ySolar = data["solar_generation_forecast"].iloc[start:stop]
yTotal = data["total_generation_forecast"].iloc[start:stop]

f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 20), sharex=True)
plt.gca().axes.get_xaxis().set_visible(False)

ax1.plot(x, yWind)
ax1.set_ylabel("wind_forecast")
ax1.set_xlabel("dateTime")

ax2.plot(x, ySolar)
ax2.set_ylabel("solar_forecast")
ax2.set_xlabel("dateTime")

ax3.plot(x, yTotal)
ax3.set_ylabel("total_forecast")
ax3.set_xlabel("dateTime")

print("Plotting from " + str(start) + " to " + str(stop) + " hour intervals = " + str((stop-start)/24) + " days.")
print("This may take some time...")
```

```python
# wind-, solar-, and total forecast error plotted vs. dateTime

yWindErrorAbs = data["wind_forecast_error"].iloc[start:stop]
ySolarErrorAbs = data["solar_forecast_error"].iloc[start:stop]
yTotalErrorAbs = data["forecast_error"].iloc[start:stop]

f, (ax4, ax5, ax6) = plt.subplots(3, 1, figsize=(20, 20), sharex=True)
plt.gca().axes.get_xaxis().set_visible(False)

ax4.plot(x, yWindErrorAbs)
ax4.set_ylabel("wind_forecast_error_absolute")
ax4.set_xlabel("dateTime")

ax5.plot(x, ySolarErrorAbs)
ax5.set_ylabel("solar_forecast_error_absolute")
ax5.set_xlabel("dateTime")

ax6.plot(x, yTotalErrorAbs)
ax6.set_ylabel("total_forecast_error_absolute")
ax6.set_xlabel("dateTime")

print("Plotting from " + str(start) + " to " + str(stop) + " hour intervals = " + str((stop-start)/24) + " days.")
print("This may take some time...")
```

```python
# wind-, solar-, and total forecast error (percent) plotted vs. dateTime

yWindErrorPerc = data["wind_forecast_error_percent"].iloc[start:stop]
ySolarErrorPerc = data["solar_forecast_error_percent"].iloc[start:stop]
yTotalErrorPerc = data["forecast_error_percent"].iloc[start:stop]

f, (ax7, ax8, ax9) = plt.subplots(3, 1, figsize=(20, 20), sharex=True)
plt.gca().axes.get_xaxis().set_visible(False)

ax7.plot(x, yWindErrorPerc)
ax7.set_ylabel("wind_forecast_error_percent")
ax7.set_xlabel("dateTime")

ax8.plot(x, ySolarErrorPerc)
ax8.set_ylabel("solar_forecast_error_percent")
ax8.set_xlabel("dateTime")

ax9.plot(x, yTotalErrorPerc)
ax9.set_ylabel("total_forecast_error_percent")
ax9.set_xlabel("dateTime")

print("Plotting from " + str(start) + " to " + str(stop) + " hour intervals = " + str((stop-start)/24) + " days.")
print("This may take some time...")
```

```python
def dfTest(colName):
    #Perform Dickey-Fuller test:
    dftest = adfuller(data[colName], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        if(dfoutput['Test Statistic'] < value):
            dfoutput['Critical Value (%s)'%key] = value
        else:
            dfoutput['Critical Value (%s)'%key] = str(value) + " :: error"
    print (dfoutput)
```

```python
# !!! solar_forecast_error_percent currently excluded due to missing values
# !!! pricediff_percent currently excluded due to missing values

columns = ['day_ahead_price', 'intraday_price', 'price_diff', 'solar_capacity', 'solar_generation_actual',
           'wind_capacity', 'wind_generation_actual', 'wind_generation_forecast', 'solar_generation_forecast',
           'total_generation_forecast', 'total_generation_actual', 'wind_forecast_error', 'solar_forecast_error',
           'forecast_error', 'forecast_error_percent', 'wind_forecast_error_percent', 
           #'solar_forecast_error_percent', 'pricediff_percent', 
           'total_capacity', 'solar_utilization', 'wind_utilization', 'total_utilization']

for col in columns:
    print("\n\n")
    print("Results of Dickey-Fuller Test for " + col + ":\n")
    dfTest(col)
```

```python
# forecastErrorSeriesData = data["forecast_error_percent"]
forecastErrorSeriesData = data[["datetime", "forecast_error_percent"]]
# forecastErrorSeriesData = forecastErrorSeriesData.astype({"datetime": datetime})
# forecastErrorSeriesData = forecastErrorSeriesData.set_index("datetime", drop=True)
```

```python
acf_function = acf(forecastErrorSeriesData["forecast_error_percent"], nlags=60) # change nlags to change number of lagging dataPoints in the past
pacf_function = pacf(forecastErrorSeriesData["forecast_error_percent"], nlags=2, method="ols")

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 20), sharex=False, sharey=False)

ax1.plot(acf_function)
ax1.set_ylim(0, 0.025) # comment this line to see the whole graph not just the intersection
ax1.set_xlim(54)       # comment this line to see the whole graph not just the intersection
ax1.set_title("Autocorrelation Function")

ax2.plot(pacf_function)
ax2.set_ylim(0, 0.025) # comment this line to see the whole graph not just the intersection
ax2.set_title("Partial Autocorrelation Function")

for ax in [ax1, ax2]:
    ax.axhline(y=0, linestyle="--", color="gray") # zero center line
    ax.axhline(y=-1.96/np.sqrt(len(forecastErrorSeriesData)), linestyle="--", color="gray") # negative confidence interval
    ax.axhline(y=1.96/np.sqrt(len(forecastErrorSeriesData)), linestyle="--", color="gray") # positive confidence interval
```

```python
dataSelection = np.asarray(forecastErrorSeriesData["forecast_error_percent"])

predictionModel = ARIMA(dataSelection, order=(2, 1, 56))
results = predictionModel.fit(disp=1)


f, (ax1) = plt.subplots(1, 1, figsize=(20, 20), sharex=False, sharey=False)

# ax1.plot(dataSelection)
ax1.plot(results.fittedvalues, color="red")

#####
# plt.title("RSS: %.4f"% sum((results.fittedvalues-forecastErrorSeriesData)**2))
```

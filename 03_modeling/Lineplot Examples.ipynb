```python
import seaborn as sns
sns.set(rc={'figure.figsize':(18,10)})
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("../00_data/data_with_calc_fields.csv", sep=";", decimal=",", index_col="datetime")
```

```python
df.index = pd.to_datetime(df.index)
df16 = df.loc['2016-01-01':'2016-12-31']
df.columns
```

```python
fig, ax = plt.subplots()

sns.lineplot(x="date", y="intraday_price", data=df16, ax=ax)

ax.set_title("Intraday prices in 2016")
ax.set_xlabel("Month of the year")
ax.set_ylabel("Intraday price")
ax.set_xticks(np.arange(360, step=30.5))
ax.set_xticklabels(["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]);
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../00_data/merged_all_with_datetime_by_algo.csv", sep=";", decimal=",", index_col=["datetime"])
data
```

```python
x = data["price_diff"]
y1 = data["solar_forecast_error"]
y2 = data["wind_forecast_error"]
y3 = data["forecast_error"]

f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 10), sharex=True)
ax1.scatter(x, y1)
ax1.set_title("Solar Forecast Error vs. Price Difference")
ax2.scatter(x, y2)
ax2.set_title("Wind Forecast Error vs. Price Difference");
ax3.scatter(x, y3)
ax3.set_title("Total Forecast Error vs. Price Difference");
```

```python
x = data["price_diff"]
x1 = data["pricediff_percent"]
y1 = data["forecast_error_percent"]
f, (ax, ax3) = plt.subplots(2, 1, figsize=(20, 10), sharex=False)
ax.scatter(x, y1)
ax.set_title("Forecast Error Percent vs. Price Difference")
ax3.scatter(x1, y1)
ax3.set_title("Forecast Error Percent vs. Price Difference Percent")
```

```python
x = data["hour"]
x1 = data["season"]
y1 = data["forecast_error"]
y2 = data["forecast_error_percent"]
f, (ax, ax1) = plt.subplots(2, 1, figsize=(20, 10), sharex=False)
ax.scatter(x, y1)
ax.set_title("Forecast Error vs. Time of day")
ax1.scatter(x1, y1)
ax1.set_title("Forecast Error vs. Season")
ax1.scatter(x1, y1)
ax1.set_title("Forecast Error Percent vs. Season")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("../00_data/data_with_calc_fields.csv", sep=";", decimal=",")
```

```python
from datetime import datetime
data_grouped = data.groupby([data['date']])
data_grouped_mean = data_grouped.mean()

x = data_grouped_mean.index
y1 = data_grouped_mean["solar_forecast_error"]
y2 = data_grouped_mean["wind_forecast_error"]

f, (ax, ax1) = plt.subplots(2, 1, figsize=(20, 20), sharex=True)
ax.scatter(x, y1, label="Solar", c="r")
ax.set_title("Solar Forecast error over whole period")

ax1.scatter(x, y2, label="Wind", c="g")
ax1.set_title("Wind Forecast error over whole period")

plt.legend()
plt.show()
```

```python
data_grouped_month = data[['month',"solar_forecast_error", "wind_forecast_error"]].groupby([data['month']])
#print(data_grouped_month[["solar_forecast_error", "wind_forecast_error"]].describe())
```

```python
data_grouped_month_mean = data_grouped_month.mean()
data_grouped_month_median = data_grouped_month.median()
data_grouped_month_min = data_grouped_month.min()
data_grouped_month_max = data_grouped_month.max()

x = data_grouped_month_mean.index
y1 = data_grouped_month_min["solar_forecast_error"]
y2 = data_grouped_month_mean["solar_forecast_error"]
y3 = data_grouped_month_median["solar_forecast_error"]
y4 = data_grouped_month_max["solar_forecast_error"]
y5 = data_grouped_month_min["wind_forecast_error"]
y6 = data_grouped_month_mean["wind_forecast_error"]
y7 = data_grouped_month_median["wind_forecast_error"]
y8 = data_grouped_month_max["wind_forecast_error"]

f, (ax) = plt.subplots(1, 1, figsize=(20, 10), sharex=True)
#ax.plot(x, y1, label="Day ahead", c="r", linestyle='--')
ax.plot(x, y2, label="Solar Mean", c="r", linestyle='--')
ax.plot(x, y3, label="Solar Median", c="r", linestyle=':')
#ax.plot(x, y4, label="Day ahead", c="r", linestyle='--')
#ax.plot(x, y5, label="Intraday", c="g", linestyle='--')
ax.plot(x, y6, label="Wind Mean", c="g", linestyle='--')
ax.plot(x, y7, label="Wind Median", c="g", linestyle=':')
#ax.plot(x, y8, label="Intraday", c="g", linestyle='--')
ax.set_title("Solar and wind forecast error through the months")
ax.axhline(y=0, color='b')
plt.legend()
plt.show()
```
